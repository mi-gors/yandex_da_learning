### Анализ оттока

<div class="alert alert-info">
<font size="3", color = "black">
Для начала посмотрим общее соотношение клиентов с оттоком и без, чтобы в дальнейшем сравнивать с ним отток по категориям.

```python
#создание функции для построения пироговой диаграммы
def pieplot(df):
    # Получение значений столбца 'churn'
    churn_counts = df['churn'].value_counts(normalize=True)
    # Создание графика
    plt.figure(figsize=(6, 6))
    plt.title('Соотношение оттока')
    plt.pie(churn_counts, autopct='%1.1f%%')
    plt.legend(title='Признак оттока', labels=('Нет', 'Есть'), loc='upper left',bbox_to_anchor=(0.9, 0.9))
    # Отображение графика
    plt.axis('equal')
    plt.show()
```

```python
pieplot(df)
```

```python
# Выбор категорий для анализа
continuous_columns = ['score', 'age', 'balance']

# Создание графиков распределения
for column in continuous_columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(data=new_df, x=column, hue='churn', kde=True,stat='density', common_norm=False)
    plt.title(f'Распределение {column} по оттоку')
    plt.xlabel(column)
    plt.ylabel('Количество')
    plt.legend(title='Признак оттока', labels=('Есть', 'Нет'))
    plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=new_df, x='est_salary', hue='churn', kde=True, common_norm=False, stat='density')
plt.title(f'Распределение зарплаты по оттоку')
plt.xlabel('Зарплата')
plt.ylabel('Плотность вероятности')
plt.xlim(0, 300000)
plt.legend(title='Признак оттока', labels=('Есть', 'Нет'))
plt.show()
```

```python
# Создание графиков для столбцов 'yaroslavl', 'ribinsk', 'rostov', 'gender', 'products' и 'credit_card'
categorical_columns = ['yaroslavl', 'ribinsk', 'rostov', 'gender', 'credit_card']
for column in categorical_columns:
    plt.figure(figsize=(8, 5))
    sns.barplot(data=df, x=column, y='churn', ci=None)
    plt.title(f'Распределение оттока по {column}')
    plt.xlabel(column)
    plt.ylabel('Доля оттока')
    plt.show()
```

## Проверка гипотез

### Средние возрасты клиентов с одним и двумя продуктами между собой между собой равны

<div class="alert alert-info">

<div class="alert alert-info">
<font size="3", color = "black">
    Гипотезы: <br/>
    Н0: Средние возрасты клиентов с одним и двумя продуктами между собой между собой равны <br/>
    Н1: Средние возрасты клиентов с одним и двумя продуктами между собой между собой не равны <br/><br/>
    Для проверки данных гипотез используем метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, так как генеральные совокупности не зависят друг от друга и именно t-критерий Стьюдента чаще всего используется при проверке равенства средних значений в двух выборках.

```python
alpha = .05

age_1 = df.query('products == 1')['age']
age_2 = df.query('products == 2')['age']
results = st.ttest_ind(age_1, age_2)

print('Средний возраст клиентов с 1 продуктом равен', age_1.mean().round(2))
print('Средний возраст клиентов с 2 продуктами равен', age_2.mean().round(2))
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

<div class="alert alert-info">
<font size="3", color = "black">
Разница в среднем возрасте между группами в 0.7 лет оказалась существенной для отвержения нулевой гипотезы.
